```python
# Now we use LabelEncoder to convert categorical data into numerical data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
home_team	=LabelEncoder()
away_team=LabelEncoder()
home_team_continent=LabelEncoder()
away_team_continent	=LabelEncoder()
```

```python
df["home_team"]=home_team.fit_transform(df["home_team"])
df["away_team"]=away_team.fit_transform(df["away_team"])
df["home_team_continent"]=home_team_continent.fit_transform(df["home_team_continent"])
df["away_team_continent"]=away_team_continent.fit_transform(df["away_team_continent"])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tournament=LabelEncoder()
city=LabelEncoder()
country	=LabelEncoder()
neutral_location=LabelEncoder()
shoot_out=LabelEncoder()
home_team_result=LabelEncoder()
```

```python
df["tournament"]=tournament.fit_transform(df["tournament"])
df["city"]=city.fit_transform(df["city"])
df["country"]=country.fit_transform(df["country"])
df["neutral_location"]=neutral_location.fit_transform(df["neutral_location"])
df["shoot_out"]=shoot_out.fit_transform(df["shoot_out"])
df["home_team_result"]=home_team_result.fit_transform(df["home_team_result"])
```

```python
df
```

```python
# Now we seperate out features(x) and Label(y) from the dataset with using iloc() function
```

```python
x=df.iloc[:,0:15]
```

```python
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
# Now we use Sciket_learn to train and test the model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# Now we use StandardScaler process to range the data between 0 and 1
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=sc.fit_transform(x_test)
```

```python
x_test
```

```python
# Now we are using DecisiontreeClassifier algorithem to predict accuracy score
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
Cf=confusion_matrix(y_test,y_pred)
```
